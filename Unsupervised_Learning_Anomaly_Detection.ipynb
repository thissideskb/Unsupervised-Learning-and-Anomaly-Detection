# Question 1
What is Dimensionality Reduction? Why is it important in machine learning?

**Answer:**  
Dimensionality Reduction is the process of reducing the number of input features in a dataset while still keeping the useful or important information. When the dataset has too many features, it becomes harder to train models, visualize patterns, and the model may overfit easily.  

By reducing the features, we make the data simpler, faster to process, and easier for machine learning models to learn from. It also helps remove noise and unnecessary information, improving the modelâ€™s performance.

# Question 2
Name and briefly describe three common dimensionality reduction techniques.

**Answer:**  
1. **PCA (Principal Component Analysis):**  
   PCA converts high-dimensional data into a smaller set of new features called principal components that capture the maximum variation in the data.

2. **t-SNE:**  
   t-SNE is mainly used for visualization. It reduces high-dimensional data into 2D or 3D form while keeping similar points close to each other.

3. **LDA (Linear Discriminant Analysis):**  
   LDA is used when labels are available. Its goal is to reduce dimensions while keeping different classes separated as clearly as possible.

# Question 4
Explain the concept of anomaly detection and its significance.

**Answer:**  
Anomaly detection is the process of identifying unusual or abnormal points in data that do not follow normal patterns. These points are called anomalies or outliers. It is important because anomalies often indicate fraud, machine failures, cyber-attacks, or sudden unusual behavior in real-world systems.  

Detecting anomalies early can prevent losses, improve safety, and help with monitoring systems more effectively.

# Question 5
List and briefly describe three types of anomaly detection techniques.

**Answer:**  
1. **Statistical Methods:**  
   These methods assume normal data follows a distribution. If a value falls far outside the expected range (mean, standard deviation), it is considered an anomaly.

2. **Distance-Based Methods:**  
   Points far from the rest of the data are marked as anomalies because they are isolated or behave differently.

3. **Machine Learning Methods:**  
   Models like Isolation Forest and One-Class SVM learn the normal pattern of data and detect anything that does not follow that pattern.

# Question 6
What is time series analysis? Mention two key components of time series data.

**Answer:**  
Time series analysis is the study of data that is recorded at regular time intervals, such as daily stock prices or monthly sales. It helps identify trends and patterns over time so we can understand behavior or make predictions.

Two key components are:  
- **Trend:** Long-term movement in the data (upward or downward).  
- **Seasonality:** A repeating pattern that occurs at fixed time intervals, like weekly, monthly, or yearly cycles.

# Question 7
Describe the difference between seasonality and cyclic behavior in time series.

**Answer:**  
Seasonality refers to patterns that repeat at fixed and predictable intervals, such as increased sales every December or more electricity usage during winter.  
Cyclic behavior does not follow a fixed schedule. It depends on conditions like economic changes or business cycles. Cycles can last for months or years but do not repeat consistently like seasonal patterns.

```python
# Question 8
# Perform K-means clustering on a sample dataset.

# Answer:
# We generate a sample dataset, apply K-Means with 3 clusters, and plot the result.

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Create sample data
X, y = make_blobs(n_samples=200, centers=3, random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)
labels = kmeans.fit_predict(X)

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.title("K-Means Clustering")
plt.show()
```

```python
# Question 9
# What is inheritance in OOP? Give an example.

# Answer:
# Inheritance allows a child class to use the methods of a parent class.

class Animal:
    def sound(self):
        return "Animal makes a sound"

class Dog(Animal):  # Dog inherits from Animal
    def bark(self):
        return "Dog barks"

d = Dog()
print(d.sound())  # inherited
print(d.bark())   # own method
```

# Question 10
How can time series analysis be used for anomaly detection?

**Answer:**
Time series analysis checks how data changes over time. An anomaly is detected when a value suddenly breaks the normal pattern.

It can detect anomalies by:
- Comparing actual values with predicted values (forecasting)
- Checking sudden spikes or drops outside the normal range
- Using moving averages to spot unusual deviations

If a point is too high or too low compared to the expected pattern, it is marked as an anomaly.
